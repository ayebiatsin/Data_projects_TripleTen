````python
# =========================
# 1. Imports necessary libraries
# =========================

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, roc_auc_score, classification_report
from sklearn.utils import resample

RANDOM_STATE = 12345
````

### 1. Import libraries

I imported the necessary libraries for the completion of our task 

````python
# =========================
# 2. Load dataset
# =========================

df = pd.read_csv('/datasets/Churn.csv')

print(df.head())
print(df.info())
````

### 2. Load Dataset

 I loaded the libray using panda and print a sample of our dataset

````python

# =========================
# 3. Preprocessing
# =========================

# Fix missing tenure if any
df['Tenure'] = df['Tenure'].fillna(df['Tenure'].median())

# Target
target = df['Exited']

# Drop useless columns
features = df.drop(columns=['Exited', 'RowNumber', 'CustomerId', 'Surname'])

# OHE ONLY geography and gender
features = pd.get_dummies(features, columns=['Geography', 'Gender'], drop_first=True)

print("Features after OHE:")
print(features.head(), "\n")
print("Columns:", features.columns, "\n")
````

### 3. Data Prepocessing 

 **Handling missing values**

The column `Tenure` contained missing values. Since this is a numeric, continuous feature, missing values can negatively affect training.  
I chose to fill missing values with the **median** because:

- The median is robust to outliers
- It preserves the central tendency of the feature
- It is appropriate when the proportion of missing values is small

**Dropping identifier columns**

I dropped the following columns:

- `RowNumber`
- `CustomerId`
- `Surname`

These are identifiers and do not contain useful signal for predicting churn. Keeping them could lead to overfitting, since the model might memorize specific IDs instead of learning general patterns.

**Encoding categorical features**

I applied **One-Hot Encoding (OHE)** to:

- `Geography`
- `Gender`

I used `pd.get_dummies(..., drop_first=True)` so that:

- The model can work with numeric features only
- We avoid imposing any artificial ordering on categories
- We reduce collinearity by dropping one level per feature (reference category)


````python
# =========================
# 4. Check class imbalance
# ========================

df['Exited'].value_counts(normalize=True)
````

### 4. Class balance analysis

Before training, I examined the distribution of the target variable `Exited`:

- `0` (customer stayed): about 80%
- `1` (customer exited): about 20%

This means the dataset is **imbalanced**.  
If we train a model directly on this data, it might:

- Learn to predict class `0` almost all the time
- Achieve high accuracy but low recall for the minority class (`1`)
- Perform poorly on the customers we actually care about: those who are likely to leave

Because of this, we will:

1. First train models on the **original imbalanced** data (baseline), then  
2. Apply techniques to **fix the imbalance** and compare performance.


````python

# =========================
# 5. Train/Valid/Test Splits
# =========================

# First split: train_valid vs test
features_train_valid, features_test, target_train_valid, target_test = train_test_split(
    features, target,
    test_size=0.20,
    stratify=target,
    random_state=RANDOM_STATE
)

# Second split: train vs valid
features_train, features_valid, target_train, target_valid = train_test_split(
    features_train_valid, target_train_valid,
    test_size=0.25,   # gives 60/20/20
    stratify=target_train_valid,
    random_state=RANDOM_STATE
)

print("Train size:", len(features_train))
print("Valid size:", len(features_valid))
print("Test size:", len(features_test), "\n")
````

### 5. Splitting the dataset into train, validation, and test sets

I split the data in two steps:

1. **Train+validation vs test**  
   - `train_test_split(..., test_size=0.20, stratify=target, random_state=12345)`  
   - 80% of the data is used for training and validation, 20% is held out as the final **test** set.  
   - I used `stratify=target` to preserve the original class ratio (approx. 80% stayed / 20% exited) in each split.

2. **Train vs validation**  
   - I split the 80% training data again into 60% train and 20% validation (`test_size=0.25` on the train_valid set).  
   - This yields an overall split of:
     - 60% train  
     - 20% validation  
     - 20% test  

The **training set** is used to fit models and tune parameters,  
the **validation set** is used to compare models and choose the best one,  
and the **test set** is only used once at the end for the final unbiased evaluation.

````python
# =========================
# 6. Scaling
# =========================

scaler = StandardScaler()
scaler.fit(features_train)

features_train_scaled = scaler.transform(features_train)
features_valid_scaled = scaler.transform(features_valid)
features_test_scaled = scaler.transform(features_test)
````

### 6. Scaling

**Feature scaling**

I used `StandardScaler` on the training data and applied the same transformation to the validation and test sets. Standardizing features:

- Centers them around 0 with unit variance
- Helps models like Logistic Regression converge better
- Ensures features are on a comparable scale so no feature dominates others numerically


````python
# =========================
# 7. Model Evaluation Function
# =========================

def evaluate(model, X_train, y_train, X_valid, y_valid, name="Model"):
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    proba = model.predict_proba(X_valid)[:, 1]

    f1 = f1_score(y_valid, preds)
    auc = roc_auc_score(y_valid, proba)

    print(f"=== {name} ===")
    print("F1:", round(f1, 3))
    print("ROC-AUC:", round(auc, 3))
    print(classification_report(y_valid, preds))
    print("=================\n")

    return f1, auc
````

````python
# =========================
# 8. Baseline Models (Imbalanced)
# =========================

log_reg = LogisticRegression(max_iter=2000, random_state=RANDOM_STATE)
rf = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=RANDOM_STATE)

print("Baseline Logistic Regression:")
evaluate(log_reg, features_train_scaled, target_train,
         features_valid_scaled, target_valid,
         name="Logistic Regression (imbalanced)")

print("Baseline Random Forest:")
evaluate(rf, features_train_scaled, target_train,
         features_valid_scaled, target_valid,
         name="Random Forest (imbalanced)")
````

### 8. Baseline model results (imbalanced data)

On the original imbalanced data:

- Logistic Regression achieved a relatively low F1 score and a moderate ROC-AUC.
- Random Forest performed better than Logistic Regression, with a higher F1 and higher ROC-AUC.

However, both models still struggled to correctly identify the minority class
(`Exited = 1`), which is expected because the training data is imbalanced.

These baseline results are our reference point before applying any
imbalance-handling techniques.


````python
# =========================
# 9. Fix imbalance using UPSAMPLING
# =========================

train_merged = pd.concat([features_train, target_train], axis=1)

majority = train_merged[train_merged['Exited'] == 0]
minority = train_merged[train_merged['Exited'] == 1]

minority_upsampled = resample(
    minority,
    replace=True,
    n_samples=len(majority),
    random_state=RANDOM_STATE
)

upsampled = pd.concat([majority, minority_upsampled])

features_train_up = upsampled.drop(columns=['Exited'])
target_train_up = upsampled['Exited']

# Scale again using upsampled data
scaler_up = StandardScaler()
scaler_up.fit(features_train_up)

features_train_up_scaled = scaler_up.transform(features_train_up)
features_valid_up_scaled = scaler_up.transform(features_valid)
features_test_up_scaled = scaler_up.transform(features_test)
````

````python
# =========================
# 10. Models on UPSAMPLED dataset
# =========================

log_reg_up = LogisticRegression(max_iter=2000, random_state=RANDOM_STATE)
rf_up = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=RANDOM_STATE)

print("Upsampled Logistic Regression:")
evaluate(log_reg_up, features_train_up_scaled, target_train_up,
         features_valid_up_scaled, target_valid,
         name="Logistic Regression (upsampled)")

print("Upsampled Random Forest:")
evaluate(rf_up, features_train_up_scaled, target_train_up,
         features_valid_up_scaled, target_valid,
         name="Random Forest (upsampled)")
````

### 10. Results on the upsampled dataset

After upsampling the minority class on the training set so that the class
distribution becomes approximately 50/50:

- Both Logistic Regression and Random Forest improved their F1 scores.
- The improvement was especially clear for recall on the minority class:
  the models detected more customers who actually exited.
- Random Forest on the upsampled data provided the best balance between
  precision and recall, with a higher F1 score than the baseline models.

This shows that upsampling helps the model focus more on the churned customers
and reduces the bias toward always predicting the majority class.


````python
# =========================
# 10b. Models with class_weight='balanced' (second imbalance method)
# =========================

log_reg_bal = LogisticRegression(
    max_iter=2000,
    random_state=RANDOM_STATE,
    class_weight='balanced'
)

rf_bal = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=RANDOM_STATE,
    class_weight='balanced'
)

print("Logistic Regression with class_weight='balanced':")
evaluate(
    log_reg_bal,
    features_train_scaled, target_train,
    features_valid_scaled, target_valid,
    name="LogReg (class_weight='balanced')"
)

print("Random Forest with class_weight='balanced':")
evaluate(
    rf_bal,
    features_train_scaled, target_train,
    features_valid_scaled, target_valid,
    name="RandomForest (class_weight='balanced')"
)
````

### 10b. Second imbalance-fixing approach: class_weight="balanced"

As a second approach to handle class imbalance, I trained models with
`class_weight='balanced'`:

- For Logistic Regression
- For Random Forest

With this setting, the model automatically gives **higher weight** to the minority
class (`Exited = 1`) and lower weight to the majority class (`Exited = 0`).

This approach does not change the data itself (like upsampling does), but
instead changes the **loss function**, penalizing mistakes on the minority
class more strongly.

Later, I compare the performance of:

- Baseline models (no imbalance handling)
- Upsampled models
- `class_weight='balanced'` models

to select the best model.


````python
# =========================
# 10c. Hyperparameter tuning for Random Forest (upsampled data)
# =========================

depth_list = [6, 8, 10, 12]
estimators_list = [50, 100, 200, 300]

best_f1 = 0
best_params = None

for depth in depth_list:
    for n in estimators_list:
        model = RandomForestClassifier(
            n_estimators=n,
            max_depth=depth,
            random_state=RANDOM_STATE
        )
        # use the upsampled & scaled train data
        f1, auc = evaluate(
            model,
            features_train_up_scaled, target_train_up,
            features_valid_up_scaled, target_valid,
            name=f"RF (upsampled) depth={depth}, n_estimators={n}"
        )
        
        if f1 > best_f1:
            best_f1 = f1
            best_params = (depth, n)

print("Best RF params on validation set:", best_params, "with F1 =", round(best_f1, 3))
````

### 10c. Hyperparameter tuning

To improve the Random Forest performance, I tuned two key hyperparameters on
the **upsampled** training data:

- `max_depth`: [6, 8, 10, 12]
- `n_estimators`: [50, 100, 200, 300]

For each combination, I trained a model on the upsampled training set and
evaluated it on the validation set using **F1 score** and **ROC-AUC**.

I then selected the combination with the highest F1 score on the validation set
as the best model configuration.  
These best parameters are used in the final model training.


### 10b. Results with class_weight="balanced"

Using `class_weight="balanced"` also improved the performance compared to the
baseline models:

- The F1 score increased relative to the imbalanced baseline,
  indicating better performance on the minority class.
- The ROC-AUC remained high, showing that the model ranks positive
  examples reasonably well.

Compared to upsampling, the class-weighted models offered an alternative
way to handle imbalance without changing the dataset itself.  
However, in my experiments, the **Random Forest with upsampling** provided
the best F1 score overall, so I chose it as the final model.


````python
# =========================
# 11. Final Model (Full Training + Upsampling)
# =========================

# Merge TRAIN + VALID
full_features = pd.concat([features_train, features_valid])
full_target = pd.concat([target_train, target_valid])

full = pd.concat([full_features, full_target], axis=1)

majority_full = full[full['Exited'] == 0]
minority_full = full[full['Exited'] == 1]

minority_full_up = resample(
    minority_full,
    replace=True,
    n_samples=len(majority_full),
    random_state=RANDOM_STATE
)

full_up = pd.concat([majority_full, minority_full_up])

X_full = full_up.drop(columns=['Exited'])
y_full = full_up['Exited']

# Scale final training data
scaler_final = StandardScaler()
scaler_final.fit(X_full)

X_full_scaled = scaler_final.transform(X_full)
X_test_scaled = scaler_final.transform(features_test)

final_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=RANDOM_STATE
)

final_model.fit(X_full_scaled, y_full)

test_preds = final_model.predict(X_test_scaled)
test_proba = final_model.predict_proba(X_test_scaled)[:, 1]

f1_final = f1_score(target_test, test_preds)
auc_final = roc_auc_score(target_test, test_proba)

print("===== FINAL TEST RESULTS =====")
print("Final F1:", round(f1_final, 3))
print("Final ROC-AUC:", round(auc_final, 3))
print(classification_report(target_test, test_preds))
print("================================")
````

Using the best Random Forest configuration found during tuning and the
upsampling strategy on the full training data (train + validation), I trained
the final model and evaluated it on the **test set**.
